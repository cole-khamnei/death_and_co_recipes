```python
recipe_ingredients = []
recipe_names = []

for recipe in death_and_co_recipe_finder.RECIPES:
    recipe_array = np.zeros(len(death_and_co_recipe_finder.GROUPS))
    for ingredient_group in recipe.ingredient_groups:
        recipe_array[death_and_co_recipe_finder.GROUPS.index(ingredient_group)] = 1
    recipe_ingredients.append(recipe_array)
    recipe_names.append(recipe.name)

recipe_ingredient_groups = pd.DataFrame(recipe_ingredients, columns=death_and_co_recipe_finder.GROUPS)
recipe_ingredient_groups["name"] = recipe_names
recipe_ingredient_groups
```

```python
import os

from collections import defaultdict

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
light_colors = {
    "Red": (255,59,48),
    "Orange": (255,149,0),
    "Yellow": (255,204,0),
    "Green": (52,199,89),
    "Mint": (0,199,190),
    "Teal": (48,176,199),
    "Cyan": (50,173,230),
    "Blue": (0,122,255),
    "Indigo": (88,86,214),
    "Purple": (175,82,222),
    "Pink": (255,45,85),
    "Brown": (162,132,94)
}
light_colors_hex = {}
for color, value in light_colors.items():
    value_hex = "#" + "".join(("0" + hex(channel).split("x")[-1])[-2:] for channel in value)
    value_hex = value_hex.upper()
    light_colors_hex[color] = value_hex

dark_colors = {
    "Red": (255,69,58),
    "Orange": (255,159,10),
    "Yellow": (255,214,10),
    "Green": (48,209,88),
    "Mint": (102,212,207),
    "Teal": (64,200,224),
    "Cyan": (100,210,255),
    "Blue": (10,132,255),
    "Indigo": (94,92,230),
    "Purple": (191,90,242),
    "Pink": (255,55,95),
    "Brown": (172,142,104)
}
dark_colors_hex = {}
for color, value in dark_colors.items():
    value_hex = "#" + "".join(("0" + hex(channel).split("x")[-1])[-2:] for channel in value)
    value_hex = value_hex.upper()
    dark_colors_hex[color] = value_hex
    
```

```python
sns.set_palette(sns.color_palette(dark_colors_hex.values()))

glass_counts = death_and_co_recipe_finder.ORIGINAL_RECIPES.groupby(by="Glass",
                                                                   axis=0).count().Name.sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(12, 5))
sns.barplot(x=glass_counts.index.tolist(), y=glass_counts.values, palette=dark_colors_hex.values())
ax.set_xticklabels(ax.get_xticklabels(), rotation = -45)
ax.set_xlabel("Glass Type")
ax.set_ylabel("Occurence")
ax.set_title("Glass Type Occurence in Death and Co. Recipes")
fig.tight_layout(rect=(0.05, 0.05, 0.95, 0.95))
fig.savefig("plots/glass_occurences.png", dpi=100)
```

```python
ingredient_counts = defaultdict(lambda : 0)

for recipe in death_and_co_recipe_finder.RECIPES:
    for ingredient_group in recipe.ingredient_groups:
        ingredient_counts[ingredient_group] += 1

ingredient_groups = list(ingredient_counts.keys())
counts = list(ingredient_counts.values())

ingredient_group_counts = pd.DataFrame({"Ingredient": ingredient_groups,
                                        "Counts": counts}).sort_values(by="Counts", ascending=False)
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
sns.barplot(data=ingredient_group_counts, x="Ingredient", y="Counts", palette=dark_colors_hex.values())
ax.set_xticklabels(ax.get_xticklabels(), rotation = -45)
ax.set_xlabel("Ingredient")
ax.set_ylabel("Occurence")
ax.set_title("Ingredient Occurence in Death and Co. Recipes")
fig.tight_layout(rect=(0.05, 0.05, 0.95, 0.95))
fig.savefig("plots/ingredient_occurences.png", dpi=100)
```
